```python
# Import Dependencies
import numpy as np 
import pandas as pd
```

```python
# Import csv data
chilis_data = pd.read_csv('Menus/Chilis.csv')
chilis = chilis_data.copy()
dataset = chilis_data.copy()
chilis_data
```

```python
# Dish Data ETL Function
def allergenETL(dataset):
    #Restaurant Name Modification
    dataset['restaurant_name'] = "Chili's"
    dataset['restaurant_id'] = 1
    dataset['dish_id'] = 0
    
    # Ensure all columns are present and are FALSE if not present
    column_names = dataset.columns.values.tolist()
   
    # Change 1.0 to TRUE
    for column in column_names:
        dataset[column] = np.where((dataset[column] == 1.0), True, dataset[column])

     # Change 1.0 to TRUE
    for column in column_names:
        if dataset[column].dtype == float:
            dataset[column] = dataset[column].astype(object)
            dataset[column] = np.where((dataset[column] == 1.0), True, dataset[column])
            dataset[column] = dataset[column].fillna(False)
            dataset[column] = dataset[column].astype(bool)

    # Define column types: 
    all_cols = ['dish_id', 'dish_name', 'restaurant_id', 'restaurant_name', 'dish_type', 'ingredients', 'modifiable', 'cross_contam', 'vegan', 'vegetarian', 'gluten', 'dairy', 'tree_nuts', 'seeds', 'nightshades', 'egg', 'fish', 'peanut', 'shellfish', 'molllusks', 'legumes', 'garlic', 'soy', 'corn', 'wheat', 'barley', 'rye', 'sesame', 'coffee', 'buckwheat', 'lupin', 'mustard', 'celery', 'oat', 'rice', 'acerloa', 'apple', 'apricot', 'banana', 'cherry',	'dates', 'fig', 'grape', 'kiwi', 'lychee', 'mango', 'melon', 'orange', 'peach', 'pear', 'persimmon', 'pineapple', 'pomegranate', 'prune', 'strawberry', 'tomato', 'chickpea', 'lentil', 'pea', 'asparagus', 'avocado', 'bell_pepper', 'cabbage', 'carrot', 'lettuce', 'potato', 'pumpkin', 'turnip', 'zucchini', 'almond', 'brazil_nut', 'cashew', 'chestnut', 'hazelnut', 'macadamia', 'pecan', 'pine_nut', 'pistachio', 'walnut', 'mustard_seed', 'poppy_seed', 'pumpkin_seed', 'sunflower_seed', 'cow_milk', 'goat_milk', 'sheep_milk', 'alaska_pollack', 'carp', 'cod', 'dogfish', 'mackeral', 'salmon', 'sole', 'tuna','crustacean', 'crab', 'lobster', 'shrimp', 'abalone', 'horned_turban', 'limpet', 'marine_snail', 'mussel', 'octopus', 'oyster', 'scallop', 'squid', 'whelk', 'beef', 'chicken', 'frog', 'honey', 'mushroom', 'mycoprotein']

    # Boolean columns
    bool_cols = ['modifiable', 'cross_contam', 'vegan', 'vegetarian', 'gluten', 'dairy', 'tree_nuts', 'seeds', 'nightshades', 'egg', 'fish', 'peanut', 'shellfish', 'molllusks', 'legumes', 'garlic', 'soy', 'corn', 'wheat', 'barley', 'rye', 'sesame', 'coffee', 'buckwheat', 'lupin', 'mustard', 'celery', 'oat', 'rice', 'acerloa', 'apple', 'apricot', 'banana', 'cherry',	'dates', 'fig', 'grape', 'kiwi', 'lychee', 'mango', 'melon', 'orange', 'peach', 'pear', 'persimmon', 'pineapple', 'pomegranate', 'prune', 'strawberry', 'tomato', 'chickpea', 'lentil', 'pea', 'asparagus', 'avocado', 'bell_pepper', 'cabbage', 'carrot', 'lettuce', 'potato', 'pumpkin', 'turnip', 'zucchini', 'almond', 'brazil_nut', 'cashew', 'chestnut', 'hazelnut', 'macadamia', 'pecan', 'pine_nut', 'pistachio', 'walnut', 'mustard_seed', 'poppy_seed', 'pumpkin_seed', 'sunflower_seed', 'cow_milk', 'goat_milk', 'sheep_milk', 'alaska_pollack', 'carp', 'cod', 'dogfish', 'mackeral', 'salmon', 'sole', 'tuna','crustacean', 'crab', 'lobster', 'shrimp', 'abalone', 'horned_turban', 'limpet', 'marine_snail', 'mussel', 'octopus', 'oyster', 'scallop', 'squid', 'whelk', 'beef', 'chicken', 'frog', 'honey', 'mushroom', 'mycoprotein'] 

    # Integer columns
    int_cols = ['dish_id', 'restaurant_id']

    # Object columns
    obj_cols = ['dish_name', 'restaurant_name', 'dish_type', 'ingredients']

    # Create the missing columns with the correct types
    for column in bool_cols:
        if column not in column_names:
            dataset[column] = False
        else: pass
    
    # for column in int_cols:
    #     if column not in column_names:
    #      dataset[column].astype(int)
    #     #  dataset[column] = pd.to_numeric(dataset[column], downcast = 'integer')
    #     #  dataset[column] = dataset[column].astype(int, errors='raise')
    #     else: pass

    for column in obj_cols:
        if column not in column_names:
            dataset[column] = ''
            dataset[column].astype(object)
        else: pass

    # # Expanding Vegetarian/Vegan
    dataset['vegetarian'] = np.where((dataset['veg/vgn'] == 'VT'), True, False)
    dataset['vegan'] = np.where((dataset['veg/vgn'] == 'VG'), True, False)
    dropthese = ['veg/vgn']
    dataset = dataset.drop(columns = dropthese, axis=1)

    # Order Columns
    dataset = dataset[all_cols]

    return dataset
```

```python
chilis = chilis_data.copy()
data = allergenETL(chilis)
data
```
